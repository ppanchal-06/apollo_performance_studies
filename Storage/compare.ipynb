```python
import pandas as pd
import matplotlib.pyplot as plt
import mplhep
import os
plt.style.use(mplhep.style.CMS)


file_paths = [
    "file1.csv",
    "file2.csv",
]

# file_paths = [
#     "serenity1/Serenity1_small.csv",
#     "serenity2/Serenity2_small.csv",
#     "serenity3/Serenity3_small.csv",
#     "serenityS1_8/SerenityS1-08_small.csv"
# ]


grouped_data = {}
for file in file_paths:
    label = os.path.splitext(os.path.basename(file))[0]
    df = pd.read_csv(file)

    df['data_size_MB'] = pd.to_numeric(df['data_size_MB'], errors='coerce')
    df['Time_s'] = pd.to_numeric(df['Time_s'], errors='coerce')
    df['speed_MBps'] = pd.to_numeric(df['speed_MBps'], errors='coerce')
    df = df.dropna(subset=['data_size_MB', 'Time_s', 'speed_MBps'])


    grouped = df.groupby('data_size_MB').agg({
        'speed_MBps': 'mean',
        'Time_s': 'mean'
    }).reset_index()

    grouped_data[label] = grouped


plt.figure(figsize=(8, 7))
for label, grouped in grouped_data.items():
    plt.plot(grouped['data_size_MB'], grouped['speed_MBps'], marker='o', label=label)

plt.xlabel('Data Size (MB)')
plt.ylabel('Bandwidth (MB/s)')
plt.ylim(0, 1000)
plt.grid(True)
plt.legend()
mplhep.cms.label("Private", data=True, rlabel="")
plt.tight_layout()
plt.savefig("compare_bandwidth_vs_datasize.pdf")
plt.show()

plt.figure(figsize=(8, 7))
for label, grouped in grouped_data.items():
    plt.plot(grouped['data_size_MB'], grouped['Time_s'], marker='o', label=label)

plt.xlabel('Data Size (MB)')
plt.ylabel('Latency (Seconds)')
plt.grid(True)
plt.legend()
mplhep.cms.label("Private", data=True, rlabel="")
plt.tight_layout()
plt.savefig("compare_latency_vs_datasize.pdf")
plt.show()
```
